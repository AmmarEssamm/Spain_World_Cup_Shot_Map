```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.ticker as ticker
import matplotlib.patheffects as path_effects
from matplotlib.patheffects import Stroke, Normal, withStroke
import matplotlib.patches as patches
from matplotlib import rcParams
from highlight_text import ax_text, fig_text
import pandas as pd
from statsbombpy import sb
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="statsbombpy")

pd.set_option('display.max_rows', None)  # To display all rows
pd.set_option('display.max_columns', None)  # To display all columns

import matplotlib.font_manager as font_manager
font_path = r"C:\Users\almagd-tec\AppData\Local\Microsoft\Windows\Fonts\Arvo-Bold.ttf"
font_manager.fontManager.addfont(font_path)
font_props = font_manager.FontProperties(fname=font_path)

from PIL import Image
import urllib
import os
```

```python
pd.set_option('display.max_rows', None)  # To display all rows 
pd.set_option('display.max_columns', None)  # To display all columns

df = sb.competitions()
df
```

```python
world_cup = sb.matches(competition_id=72, season_id=107)
world_cup
```

```python
spain_matches = world_cup[(world_cup['home_team']== "Spain Women's") | (world_cup['away_team'] == "Spain Women's")]
spain_matches
```

```python
df_spain = pd.DataFrame()
match_ids= (spain_matches.match_id.values)
for match in match_ids:
    match_data = sb.events(match_id=match)
    shots= match_data[match_data['type']== 'Shot']
    df_spain = pd.concat([df_spain, shots])
```

```python
df_spain[['end_x','end_y','end_z']] = df_spain.shot_end_location.apply(pd.Series)
df_spain
```

```python
shots_spain = df_spain[(df_spain['team']== "Spain Women's") & (df_spain['shot_outcome'] != 'Off T') & (df_spain['end_z'].isna() == False)]
shots_spain.shape
```

```python
shots_spain[['end_y','end_z']].describe()
1
```

```python


shots_spain.shapefig, ax = plt.subplots(figsize=(16,9)) 

width = 24
height = 8

increment_x = int(width / 6)
increment_y = int(height / 2)

shots_spain['goalMouthY'] = shots_spain['end_y'] - 35.7  # Subtract min value of 'end_y'
shots_spain['goalMouthY'] = [(x*width) / 9.9 for x in shots_spain['goalMouthY']]  # Scale 'goalMouthY' based on width

shots_spain['goalMouthZ'] = [(x*height) / 3.0 for x in shots_spain['end_z']]  # Scale 'goalMouthZ' based on height

# Output the transformed columns
shots_spain[['goalMouthY', 'goalMouthZ']]


#plot

ax.plot([0,0],[0,height], color = 'black', lw = 3)
ax.plot([width,width],[0,height], color = 'black', lw = 3)

ax.plot([0,width],[height,height], color = 'black', lw = 3)

# Parting_zones

bins_y = range(0, width + 1, increment_x)
bins_z = range(0, height + 1, increment_y)
shots_spain['bins_y'] = pd.cut(shots_spain['goalMouthY'], bins_y)
shots_spain['bins_z'] = pd.cut(shots_spain['goalMouthZ'], bins_z)
shots_spain['shot_aux'] = 1


data_bins = shots_spain[['team','bins_y','bins_z','shot_aux']].groupby(['bins_y','bins_z','team'])['shot_aux'].sum().reset_index()
total_shots = shots_spain.groupby(['team'])['shot_aux'].sum().reset_index()
total_shots.columns = ['team','total']
data_bins = pd.merge(data_bins,total_shots, how = 'left', on = 'team')
data_bins['shot_pct'] = data_bins['shot_aux']/data_bins['total']
data_bins = data_bins.sort_values(by = ['bins_y', 'bins_z'])


data = list(data_bins['shot_pct'])
max_data = max(data)
scaled_data = [x/max_data for x in data]


i = 0
x = 0


while x < width:
    for y in range(0, height, increment_y):
        rect = patches.Rectangle(
                (x, y),  # bottom left starting position (x,y)
                increment_x,  # width
                increment_y,  # height
                ec='#ab2a3e',
                fc='#ab2a3e',
                alpha = scaled_data[i], # <---- the transparency
                zorder=-1
                )
    
        ax.add_patch(rect)
    
        # -- Anotate the counter (i) and choose color depending on value
        if scaled_data[i] < .5:
            color_text = 'black'
            fore_color ='white'
        else:
            color_text = 'white'
            fore_color = 'black'
        label_ = ax.text(
                    x = x + increment_x/2, y = y + increment_y/2,
                    s = f'{data[i]:.1%}', # <----- the data label
                    color = color_text,
                    va = 'center',
                    ha = 'center',
                    size = 13.5
                )
        # Set path effects to ensure readability
        label_.set_path_effects([Stroke(linewidth=1.7, foreground=fore_color), Normal()])
    
        i += 1
    
    # -- Once we've placed the top & bottom rectangles we move right.
    x = x + increment_x

ax.set_xlim(-2,width + 2)
ax.set_ylim(-2,height + 3)
ax.set_axis_off()
ax.plot([-2,width + 2],[0,0], color = 'black', marker = 'None', lw = 3, zorder = 3) 
title = ax.set_title("Spain Women's World Cup Shot Map:\nZones and Frequencies", 
                     fontproperties=font_props, 
                     fontsize=24, 
                     loc='center')

# Adjust title position by setting the bbox
title.set_position([0.5, 0.55]) 

ax.scatter(shots_spain['goalMouthY'], shots_spain['goalMouthZ'], color = "#ab2a3e", alpha = 0.4, ec='black')
goal_shots_spain = shots_spain[shots_spain['shot_outcome'] == 'Goal']
ax.scatter(goal_shots_spain['goalMouthY'], goal_shots_spain['goalMouthZ'], ec = "black", color = "#ab2a3e", alpha = 0.99, lw=1.15, s=400)

plt.savefig('Spain_World_Cup_Shot_Map.png', dpi=300, bbox_inches='tight')
```
